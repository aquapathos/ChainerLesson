# 3層NNによるMNIST文字認識   with Chainer

```python
import chainer
import chainer.functions as F
import chainer.links as L
from chainer import training, Chain
import chainer.optimizers as optimizers
from chainer.training import extensions as ext
import  matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
```

```python
batchsize = 50
epoch = 10
logdir = "chlog"
resume = ''
```

```python
# MNIST dataset　の読み込み
train, test = chainer.datasets.get_mnist()
# イタレータの設定
train_iter = chainer.iterators.SerialIterator(train, batchsize)
test_iter = chainer.iterators.SerialIterator(test, batchsize,
                                                 repeat=False, shuffle=False)
```

chainer の場合、train, test は、（1次元化 numpy配列としての画像, 正解番号）という対のリストである
正解を与える際に[0,1,0,0,...]  のような表現に直す必要はない。

```python
# ネットワークの定義
class MLP(Chain):
    def __init__(self,xnum, n1,n2,nclass):
        super(MLP, self).__init__()
        with self.init_scope():
            self.l1 = L.Linear(xnum, n1)  # n_in -> n_units
            self.l2 = L.Linear(None, n2)  # n_units -> n_units
            self.l3 = L.Linear(None, nclass)  # n_units -> n_out

    def __call__(self, x):
        h1 = F.relu(self.l1(x))
        h2 = F.relu(self.l2(h1))
        y = self.l3(h2)
        return y
```

```python
model = L.Classifier(MLP(784,100,100,10))
```

- [chainer.links.Classifier](http://docs.chainer.org/en/stable/reference/generated/chainer.links.Classifier.html0

Classifier は　MNISTに代表されるようなシンプルな分類器を意味する。このタイプは 暗黙値が
- 出力の特製関数として softmax を用い、出力は一人勝ちの確率分布となる
- 目的関数は交差エントロピーの総和 → loss
- 正解率は → accuracy
となっている。ブラックボックス化されているのでこれらは表面に見えないが、それぞれ  model.lossfun,  model.loss, model.accuracy として参照可能である。

```python
# 最適化器の設定
optimizer = optimizers.Adam()
optimizer.setup(model)
```

```python
# トレーナーの設定
updater = training.StandardUpdater(train_iter, optimizer)
trainer = training.Trainer(updater, (epoch, 'epoch'), out=logdir)
```

```python
# テスト用データの評価を追加（オプション）
trainer.extend(ext.Evaluator(test_iter, model))

# 計算グラフの出力（オプション）
trainer.extend(ext.dump_graph('main/loss'))

# スナップショットの記録（オプション）
trainer.extend(ext.snapshot(), trigger=(100, 'epoch'))

# 統計データのログ（オプション）
trainer.extend(ext.LogReport())

# lossとaccuracyのグラフの出力
if ext.PlotReport.available():
        trainer.extend(
            ext.PlotReport(['main/loss', 'validation/main/loss'],
                                  'epoch', file_name='loss.png'))
        trainer.extend(
            ext.PlotReport(
                ['main/accuracy', 'validation/main/accuracy'],
                'epoch', file_name='accuracy.png'))

# 標準出力への統計データの出力
trainer.extend(ext.PrintReport(
        ['epoch', 'main/loss', 'validation/main/loss',
         'main/accuracy', 'validation/main/accuracy', 'elapsed_time']))

# プログレスバーの表示
# trainer.extend(ext.ProgressBar())

# 再開モードの場合
if resume:
    # Resume from a snapshot
    chainer.serializers.load_npz(resume, trainer)
```

```python
# 学習の実行
trainer.run()
```

```python
def plotchr(image,label,i):
    sns.set_context("talk")
    plt.subplot(1,6,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title("%d" % label)
    plt.imshow(image,cmap=plt.cm.gray_r)

for i in range(6):
    image, label = train[i]
    image = image.reshape((28,28))
    plotchr(image,label,i)
```
